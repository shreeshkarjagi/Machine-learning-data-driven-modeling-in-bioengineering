#### (1) Write a function that stores the first 1000 numbers of the Fibonacci sequence in a numpy array.

Then write two assertions as tests of your function.

```python
def fibonacci_1000():
    getcontext().prec = 1000
    fib_list = [Decimal(0), Decimal(1)]
    for i in range(2, 1001):
        next_term = fib_list[i-1] + fib_list[i-2]
        fib_list.append(next_term)
    return np.array(fib_list)

fib1000 = fibonacci_1000()
# print(fib1000)
assert fib1000[10] == Decimal('55')
assert fib1000[100] == Decimal('354224848179261915075')
assert fib1000[900] == Decimal('54877108839480000051413673948383714443800519309123592724494953427039811201064341234954387521525390615504949092187441218246679104731442473022013980160407007017175697317900483275246652938800')
assert fib1000[999] == Decimal('26863810024485359386146727202142923967616609318986952340123175997617981700247881689338369654483356564191827856161443356312976673642210350324634850410377680367334151172899169723197082763985615764450078474174626')
assert fib1000[1000] == Decimal('43466557686937456435688527675040625802564660517371780402481729089536555417949051890403879840079255169295922593080322634775209689623239873322471161642996440906533187938298969649928516003704476137795166849228875')

```
